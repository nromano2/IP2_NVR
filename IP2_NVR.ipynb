**Column Data**

```python
df.columns
```

```python
orig_column_names = df.columns
```

```python
alt_column_names = ['date_time',
                    'temp_c',
                    'dew_point',
                    'relative_humidity',
                    'wind_speed',
                    'visibility',
                    'pressure',
                    'weather']
```

```python
df.columns = alt_column_names
```

```python
df.describe()
```

```python
df['temp_f'] = ((9/5)*df['temp_c'])+32
```

```python
df = df[['date_time',
         'temp_f',
         'temp_c',
         'dew_point',
         'relative_humidity',
         'wind_speed',
         'visibility',
         'pressure',
         'weather']]
```

```python
df.head(5)
```

```python
df.dtypes
```

```python
df.date_time = pd.to_datetime(df.date_time)
```

```python
df.dtypes
```

**Monthly Averages**

```python
month_averages = df['temp_f'].groupby(df["date_time"].dt.month).mean()
```

```python
month_averages
```

```python
monthly_avg = pd.DataFrame(month_averages)
```

```python
monthly_avg
```

```python
monthly_avg['month'] = monthly_avg.index
```

```python
monthly_avg
```

```python
monthly_avg = monthly_avg[['month','temp_f']]
```

```python
monthly_avg.rename_axis(index = {'date_time':''}, inplace = True)
```

```python
monthly_avg
```

```python
month_names = {1: 'January',
               2: 'Feburary',
               3: 'March',
               4: 'April',
               5: 'May',
               6: 'June',
               7: 'July',
               8: 'August',
               9: 'September',
               10: 'October',
               11: 'November',
               12: 'December'}
```

```python
def month_conversion(x):
    return month_names.get(x)
```

```python
monthly_avg['month']=monthly_avg.month.map(month_conversion)
```

```python
monthly_avg
```

**Hourly Averages**

```python
hourly_avg = df['temp_f'].groupby(df["date_time"].dt.hour).mean()
```

```python
hourly_avg
```

```python
hourly_avg.rename_axis('Hour', inplace=True)
```

```python
hourly_avg
```

```python
time = {0: '1:00 am',
        1: '2:00 am',
        2: '3:00 am',
        3: '4:00 am',
        4: '5:00 am',
        5: '6:00 am',
        6: '7:00 am',
        7: '8:00 am',
        8: '9:00 am',
        9: '10:00 am',
        10: '11:00 am',
        11: '12:00 pm',
        12: '1:00 pm',
        13: '2:00 pm',
        14: '3:00 pm',
        15: '4:00 pm',
        16: '5:00 pm',
        17: '6:00 pm',
        18: '7:00 pm',
        19: '8:00 pm',
        20: '9:00 pm',
        21: '10:00 pm',
        22: '11:00 pm',
        23: '12:00 am'}
```

```python
def hour_conversion(x):
    return time.get(x)
```

```python
hourly_avg.index = hourly_avg.index.map(hour_conversion)
```

```python
hourly_avg
```

**Weather Data**

```python
df['weather'].groupby(df['weather']).count()
```

```python
types = df.weather.str.split(pat = ',', expand=True)
weather_types = set(types[0].unique()).union((set(types[1].unique())).union(set(types[2].unique())))
print(weather_types)
```

```python
weather_types_list = list(weather_types)
weather_types_list.remove(None)
weather_types_list.sort()
weather_types_list
```

```python
for i in weather_types_list:
    i = str(i)
    print(f"{i}: {df['weather'].str.contains(i).sum()}")
```
